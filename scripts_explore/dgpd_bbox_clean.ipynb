# Adding bounding box columns & cleaning out global scope rows

Dask probably not strictly necessary for the 4 million row file used here/for such row-wise operations

```python
ddf = dgp.read_parquet("merged_sorted_ddb.parquet", gather_spatial_partitions=False)
```

```python
ddf.head()
```

```python
ddfb = ddf.bounds
```

```python
ddfb.head()
```

```python
gdf = dd.concat([ddf, ddfb], axis=1)
```

```python
gdf.head()
```

```python
# Pull gdf rows where minx, maxx, miny and maxy are not -180, -90, 180, or 90
bbox_columns = ['minx', 'maxx', 'miny', 'maxy']

# Define the values to check against
check_values = [-180, -90, 90, 180]

# Create a boolean mask for rows where any of the bbox columns are in check_values
mask = gdf[bbox_columns].isin(check_values).all(axis=1)

# Count the number of rows that match the condition
count = mask.sum().compute()

print(f"Number of rows with bounding box values in {check_values}: {count}")
```

```python
# If you want to see these rows, you can use:
rows_of_interest = gdf[mask].compute()
print("\nFirst few rows that match the condition:")
print(rows_of_interest[bbox_columns].head())

# If you want to see the unique values in these columns:
for col in bbox_columns:
    unique_values = gdf[mask][col].unique()
    print(f"\nUnique values in {col} (excluding {check_values}):")
    print(unique_values)
```

```python
# negative version of mask 
maskneg = ~gdf[bbox_columns].isin(check_values).all(axis=1)
```

```python
cleaned_gdf = gdf[maskneg].compute()
```

```python
row_count = cleaned_gdf.shape[0]
print(f"Number of rows in cleaned DataFrame: {row_count}")
```

```python
cleaned_gdf.head()
```

```python
output_file = '3mil_no_global_bounds.parquet'
```

```python
# write the cleaned data to new file
cleaned_gdf.to_parquet(output_file, row_group_size = 100000, compression='snappy', index=False)
```
